```python
from skimage import data, segmentation, color
from skimage.future import graph
from matplotlib import pyplot as plt
import cv2 as cv2
import numpy as np
```

```python
img = cv2.imread('Dataset/camera4/JPEGImages/00001.jpg')
img = cv2.resize(img, (224,224))
img  = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
labels1 = segmentation.slic(img, compactness=10, n_segments=350)
out1 = color.label2rgb(labels1, img, kind='avg')
g = graph.rag_mean_color(img, labels1, mode='similarity')
labels2 = graph.cut_normalized(labels1, g, thresh=0.01)
out2 = color.label2rgb(labels2, img, kind='avg')

plt.imshow(segmentation.mark_boundaries(out1, labels1))
plt.savefig("slic.png")
plt.show()
plt.imshow(out2)
plt.savefig("ncut.png")
plt.show()
```
